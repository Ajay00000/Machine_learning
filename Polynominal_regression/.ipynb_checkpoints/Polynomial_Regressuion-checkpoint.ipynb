```python
import matplotlib.pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

##### Creating Data set With curv linier 

```python
x = 10*np.random.normal(0,1,70)
y = 10*(-x**2)+np.random.normal(-100,100,70)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=15)
plt.title("Polynomial Regression")
plt.xlabel("Predictor",fontsize=16)
plt.ylabel("Target",fontsize=16)
plt.show()
```

#### Creating Linier model

```python
model = LinearRegression()
model.fit(x.reshape(-1,1),y.reshape(-1,1))
```

##### SEE linier regression Result 

```python
ypred = model.predict(x.reshape(-1,1))
plt.figure(figsize=(10,5))
plt.scatter(x,y,s=15)
plt.xlabel("Predictor",fontsize=16)
plt.ylabel("Target",fontsize=16)
plt.plot(x,ypred,color='red')
plt.show()
```

```python
print("RMSE for Linier Regression ",np.sqrt(mean_squared_error(y,ypred)))
```

### Creating Pipeline

```python
pipe = Pipeline([('pipeline',PolynomialFeatures(degree=2)),('Regression',LinearRegression())])
pipe.fit(x.reshape(-1,1),y.reshape(-1,1))
```

```python
polypred = pipe.predict(x.reshape(-1,1))
sortedzip = sorted(zip(x,polypred))
x_poly, poly_pred = zip(*sortedzip)
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=15)
plt.plot(x,ypred,color='r',label='Linear Regression')
plt.plot(x_poly,poly_pred,color='g',label='Polynomial Regression')
plt.xlabel('Predictor',fontsize=16)
plt.ylabel('Target',fontsize=16)
plt.legend()
plt.show()
```
